**Задание 2**. Сгенерировать 1000 случайных чисел с любым известным распределением
и построить их гистограмму, математическое ожидание, дисперсию.

```python
data = np.random.normal(loc = 10, scale = 2, size = 1000)
data
```

```python
plt.hist(data)
```

```python
print("Мат. ожидание = {}; Дисперсия = {}".format(np.mean(data), np.std(data)))
```

```python
a = 5
matrix = np.random.uniform(low = 0, high = a, size = (1,2))
matrix
```
